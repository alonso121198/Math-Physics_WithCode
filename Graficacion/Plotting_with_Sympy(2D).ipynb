# plot Class

Son clases definidos con **Plot(*arg, **kwargs)** .Permite graficar Sympy expresiones . es una funcion de una vaiable y retorna una instancia de **Plot** Veamos como definir los objetos e interactuar con ellos . 
- plot(expr, range, ** kwargs) Si el rango no es especificado por defalut es (-10, 10 ) . 
- plot(expr1, expr2, expr3 , expr4, expr5, range, ** kwargs ) 
- plot( (expr1, range1) , (expr2, range) ,... , ** kwargs ) 
- arguments

    - range : un 3-tuple (x,min, max )

    - show(BOOLEAN) : para mostrar o no la grafica

### Examples

```python
from sympy import symbols # para simbolos
from sympy.plotting import plot # para dibujar (ver que lleva letra minuscula)
x = symbols('x')
```

```python
plot(x**2, (x,-5,5) )
```

```python
plot(x, x**2, x**3, (x,-5,5))
```

```python
plot((x**2,(x,-6,6)), (x,(x,-5,5)))
```

```python
plot(x**2, adaptive=False, nb_of_points=400)
```

## append(arg)
Aade un elemento de una serie plot en un existente plot . Esto se entendera con los ejemplos

```python
p1 = plot(x*x, show= False) # ese show es para que no se muestre
p2 = plot(x, show=False) 
 # agrega el primer objeto de la serie del segundo grafico al primero . 
p1.append(p2[0])
p1
```

```python
p1.show()
```

## extend(arg)
Agrega todas las series de otro plot . Funciona similar al anterior . con un ejemplo veras la diferencia

```python
p1 = plot(x**2, show=False)
p2 = plot(x, -x, show= False)
# aande el segundo plot al primero
p1.extend(p2)
p1
```

```python
p1.show()
```

## plot_parametric
Traza un diagrama paramtrico 2D
- plot_parametric(expr_x, expr_y , range , ** kwargs) Si no se especifica el rango por default es (-10, 10 ) 
- plot_parametric( (expr1_x, expr_1_y), (expr2_x, expr2_y) , range, ** kwargs) Multiples graficados
- Multiples graficados con distintos rangos :plot_parametric( (expr_x, expr_y, range) , .... , ** kwargs ) 
- Arguments
    - expr_x : Funcion a lo largo de x 
    - expr_y : funcion a lo largo de y
    - range : (u,0,5), un de las variables . Recuerda se esta graficando una funcion $ R \rightarrow RxR $
- Keyword Arguments
    - adaptite(BOOLEAN)
    - depth(int)
    - nb_of_points(int)
    - line_color
    - xlabel(str)
    - ylabel(str)
    - xscale
    - yscale
    - axis_center(tupla)
    - xlim(tupla)
    - ylim(tupla)

```python
from sympy import symbols, cos, sin
from sympy.plotting import plot_parametric
u = symbols('u')
```

```python
plot_parametric(cos(u), sin(u), (u,-5,5) )
```

```python
# ojo no confundas el rango de u con el rango de "x" y "y" (OJO)
plot_parametric( ( cos(u), sin(u)) , ( u,cos(u) ) )
```

```python
plot_parametric( ( cos(u), sin(u), (u,-5,5) ), (cos(u), u, (u,-5,5)))
```
