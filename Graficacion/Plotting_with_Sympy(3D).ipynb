# plot3d
**plot3d(* args, ** kwargs)** grafica en 3D
- plot3d( expr, range_x , range_y, ** kwargs) 
- plot3d( expr1, expr2, range_x, range_y, ** kwargs)
- Arguments
    - expr_x
    - range_x 
    - range_y
- Keyword Arguments
    - nb_of_points_x(int) 
    - nb_of_points_y(int) 
    - surface_color
    - title(str)
    - ...(ver la funcion plot) 

```python
from sympy import symbols
from sympy.plotting import plot3d
x,y = symbols('x y')
```

```python
plot3d(x*y, (x,-5,5),(y,-5,5) )
```

```python
plot3d(x*y, -x*y, (x,-5,5), (y,-5,5) )
```

```python
plot3d( (x**2+ y**2, (x, -5,5), (y,-5,5) ), (x*y, (x, -3,3), (y,-3,3) )) 
```

# pot3d_parametric_line
Este es igual que el de 2D . Asi que directo nos iremos a los ejemplo. Los arguments y Kewywords arguments son similares a los del 2d . Esta vez se grafica una funcion $ R \rightarrow R^3 $

```python
from sympy import symbols, cos, sin
from sympy.plotting import plot3d_parametric_line
u = symbols('u')
```

```python
plot3d_parametric_line( cos(u), sin(u), u, (u,-5,10) )
```

```python
plot3d_parametric_line( (cos(u), sin(u),u, (u,-5,5) ), (sin(u), u**2, u, (u,-5,5) ) )
```

# plot3d_parametric_surface
Dibuja una superficie para metrica en 3D . En este caso se tiene una funcion $ R^2 \rightarrow R^3 $ . 
Como los anteriores se aprende a como usarlo viendo los ejemplos . No hay mucho de nuevo que los anteriores funciones aqui . 

```python
from sympy import symbols, cos, sin
from sympy.plotting import plot3d_parametric_surface
u, v =symbols('u v')
```

```python
plot3d_parametric_surface( cos(u+v), sin(u-v), u-v, (u,-5,5), (v,-5,5) )
```

# plot_implicit
**plot_implicit(expr, x_var= None, y_var= None , adaptive=True, depth=0, points=300 , line_color='blue', show=True, ** kwargs)** es una funcion para graficar ecuaciones/desigualdades implicitas
- arguments
    - expr: la exuacion/inecuacion que va a ser graficada
    - x_var : optional (symbols, xmin, xmax)
    - y_var : optional (symbols, ymin, ymax)
- Keyword arguments
    - adaptive
    - depth
    - points
    - show
    - title(string)
    - xlabel(string)
    - ylabel(string)
    - line_color(float o string)

```python
from sympy import plot_implicit, cos, sin , symbols, Eq , And
x, y = symbols('x y')
# sin algun rango en symbolos en la expresion
p1 = plot_implicit(Eq(x**2+y**2, 5)) # fijate que grafica una funcion implicita
```

```python
# con rango para los simbolos
p2 = plot_implicit(Eq(x**2+y**2, 3), (x,-3,3), (y,-3,3)) 
```

```python
# con el argumento depth=2 , ese argumento va de 1 a 4 
# (la verdad que no se muy bne para que sirve)
p4 = plot_implicit( Eq(x**2+y**2,5), (x,-4,4), (y,-4,4), depth = 2  )
```

```python
p4 = plot_implicit(Eq(x**2+y**2,5), (x,-5,5),(y,-2,2), adaptive=False)
```

```python
# ten cuidado con ese numero points ,usalo < 1000 . Tu maquina podria explotar
# te lo ADVIERTO
p4 = plot_implicit(Eq(x**2+y**2,5), (x,-5,5),(y,-2,2), adaptive=False, points=400)
```

```python
#  graficando regiones
p6 = plot_implicit(y>x**2)
```

```python
# graficando usando conjunciones booleanas
p7 = plot_implicit(And(y>x, y>-x))
```

```python
# cuando estes graficando una expresion con una sola variable . especifica
# la variable "x"o "y" . Son buenas practicas 
p8 = plot_implicit(y-1, y_var=y)
p9 = plot_implicit(x-1, x_var=x)
```

# PlotGrid Class
**PlotGrid( nrows, ncolumns, *args, **kwargs) Esto sirve para trazar subtramas de graficos ya realizados 

```python
from sympy import symbols
from sympy.plotting import plot, plot3d, PlotGrid
x, y = symbols('x y')
p1 = plot(x, x**2, x**3,(x,-5,5))
p2 = plot( ( x**2, (x,-6,6) ), (x, (x,-5,5) ) )
p3 = plot( x**3, (x,-5,5)) 
p4 = plot3d( x*y, (x, -5,5) ,(y, -5,5) )
```

```python
# ploting verticalmente en una linea
PlotGrid(2,1,p1, p2)
```

```python
# Plotting horizontal
PlotGrid(1,3,p2,p3,p4)
```

```python
PlotGrid(2,2, p1,p2,p3,p4)
```
